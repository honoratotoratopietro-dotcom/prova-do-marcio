```python
#Solicite ao usuário um número inteiro positivo.
#Utilizando um laço de repetição, calcule e exiba a soma de todos os números de 1 até o número informado.


num = int(input("Digite um número inteiro positivo: "))

if num < 1:
    print("Por favor, digite um número inteiro positivo.")
else:
    soma = 0
    for i in range(1, num + 1):
        soma += i
    print(f"Resultado: {soma}")
```

```python
#Solicite ao usuário um número inteiro e exiba sua tabuada de 1 até 10.


num = int(input("Digite um número: "))

print(f"Tabuada do {num}:")
for i in range(1, 11):
    resultado = num * i
    print(f"{num} x {i} = {resultado}")
```

```python
#Solicite ao usuário um número inteiro positivo e calcule seu fatorial utilizando um laço de repetição.


num = int(input("Digite um número inteiro positivo: "))

if num < 0:
    print("O fatorial não é definido para números negativos. Por favor, digite um número inteiro positivo.")
elif num == 0:
    print("Fatorial = 1") # Fatorial de 0 é 1 por definição
else:
    fatorial = 1
    for i in range(1, num + 1):
        fatorial *= i
    print(f"Fatorial = {fatorial}")
```
