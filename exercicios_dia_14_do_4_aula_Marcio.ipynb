```python
#1
#Par ou ímpar
#Peça um número ao usuário e informe se ele é par ou ímpar.

numero = int(input("Digite um número: "))

if numero % 2 == 0:
    print(f"O número {numero} é PAR.")
else:
    print(f"O número {numero} é ÍMPAR.")
```

```python
#Maior de dois números
#Solicite dois números e mostre qual é o maior (ou se são iguais).


numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))

if numero1 > numero2:
    print(f"O primeiro número ({numero1}) é maior que o segundo número ({numero2}).")
elif numero2 > numero1:
    print(f"O segundo número ({numero2}) é maior que o primeiro número ({numero1}).")
else:
    print(f"Os números são iguais ({numero1}).")
```

```python
#Aprovação de aluno
#Peça a nota de um aluno e informe:

nota = float(input("Digite a nota do aluno: "))

if nota >= 7:
    print("Aprovado")
elif 5 <= nota < 7:
    print("Recuperação")
else:
    print("Reprovado")
```

```python
#Classificação de idade
#Receba a idade e classifique:

idade = int(input("Digite a idade: "))

if 0 <= idade <= 12:
    print("Criança")
elif 13 <= idade <= 17:
    print("Adolescente")
elif 18 <= idade <= 59:
    print("Adulto")
elif idade >= 60:
    print("Idoso")
else:
    print("Idade inválida")
```

```python
#Senha correta
#Defina uma senha fixa no código. Peça ao usuário para digitar e informe se está correta ou incorreta.

senha_correta = "senha123"
senha_digitada = input("Digite a senha: ")

if senha_digitada == senha_correta:
    print("Senha correta")
else:
    print("Senha incorreta")


```

```python
#Positivo, negativo ou zero
#Peça um número e diga se ele é positivo, negativo ou zero.

numero = float(input("Digite um número: "))

if numero > 0:
    print("Positivo")
elif numero < 0:
    print("Negativo")
else:

    print("Zero")



```

```python
#Calculadora simples
#Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o resultado.

num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))
operacao = input("Digite a operação (+, -, *, /): ")

if operacao == '+':
    resultado = num1 + num2
elif operacao == '-':
    resultado = num1 - num2
elif operacao == '*':
    resultado = num1 * num2
elif operacao == '/':
    if num2 != 0:
        resultado = num1 / num2
    else:
        resultado = "Erro: Divisão por zero!"
else:
    resultado = "Operação inválida!"

print(f"Resultado: {resultado}")
```

```python
#Dia da semana
#Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).

numero_dia = int(input("Digite um número de 1 a 7: "))

dias_da_semana = {
    1: "Domingo",
    2: "Segunda-feira",
    3: "Terça-feira",
    4: "Quarta-feira",
    5: "Quinta-feira",
    6: "Sexta-feira",
    7: "Sábado"
}

if numero_dia in dias_da_semana:
    print(f"{numero_dia} = {dias_da_semana[numero_dia]}")
else:
    print("Número inválido! Digite um número entre 1 e 7.")
```

```python
#Ano bissexto
#Solicite um ano e informe se ele é bissexto ou não.

ano = int(input("Digite um ano: "))

if (ano %
    4 == 0 and ano % 100 != 0) or (ano %
                                       400 == 0):
    print(f"{ano} é um ano bissexto.")
else:
    print(f"{ano} não é um ano bissexto.")

```

```python
#Desconto em compras
#Peça o valor de uma compra:
# ≥ 100: 10% de desconto
#	≥ 50 e < 100: 5% de desconto
# < 50: sem desconto

valor_compra = float(input("Digite o valor da compra: "))

if valor_compra >= 100:
    desconto = 0.10  # 10% de desconto
elif valor_compra >= 50:
    desconto = 0.05  # 5% de desconto
else:
    desconto = 0.0   # Sem desconto

valor_final = valor_compra * (1 - desconto)

print(f"Valor original: R$ {valor_compra:.2f}")
print(f"Desconto aplicado: {desconto * 100:.0f}%")
print(f"Valor final com desconto: R$ {valor_final:.2f}")
```
