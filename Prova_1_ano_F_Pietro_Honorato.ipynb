```python
#Pietro Honorato 1 ano F
# Crie um programa que leia a idade de uma pessoa e classifique conforme as regras abaixo:
	#Até 17 anos: Menor de idade
	#De 18 a 59 anos: Adulto
	#60 anos ou mais: Idoso

idade_str = input("Por favor, digite a idade da pessoa: ")

try:
    idade = int(idade_str)

    if idade <= 17:
        classificacao = "Menor de idade"
    elif 18 <= idade <= 59:
        classificacao = "Adulto"
    else:
        classificacao = "Idoso"

    print(f"A pessoa com {idade} anos é classificada como: {classificacao}")
except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro para a idade.")
```

```python
#Crie um programa que leia o valor de uma compra e aplique o desconto conforme as regras:
	#Até R$100: sem desconto
	#	De R$101 a R$500: 10% de desconto
	#	Acima de R$500: 20% de desconto

valor_compra_str = input("Por favor, digite o valor da compra: R$")

try:
    valor_compra = float(valor_compra_str)
    desconto_percentual = 0.0

    if valor_compra <= 100:
        desconto_percentual = 0.0 # Sem desconto
    elif 101 <= valor_compra <= 500:
        desconto_percentual = 0.10 # 10% de desconto
    else:
        desconto_percentual = 0.20 # 20% de desconto

    valor_desconto = valor_compra * desconto_percentual
    valor_final = valor_compra - valor_desconto

    print(f"Valor da compra: R${valor_compra:.2f}")
    print(f"Desconto aplicado: {desconto_percentual * 100:.0f}%")
    print(f"Valor do desconto: R${valor_desconto:.2f}")
    print(f"Valor final com desconto: R${valor_final:.2f}")

except ValueError:
    print("Entrada inválida. Por favor, digite um número válido para o valor da compra.")
```

```python
#Crie um programa que leia três valores e verifique se eles podem formar um triângulo.
#Para isso, a soma de dois lados deve ser maior que o terceiro.
#Mostre uma mensagem informando se é ou não um triângulo válido.

lado1_str = input("Por favor, digite o valor do primeiro lado do triângulo: ")
lado2_str = input("Por favor, digite o valor do segundo lado do triângulo: ")
lado3_str = input("Por favor, digite o valor do terceiro lado do triângulo: ")

try:
    lado1 = float(lado1_str)
    lado2 = float(lado2_str)
    lado3 = float(lado3_str)

    if lado1 > 0 and lado2 > 0 and lado3 > 0:
        # Condição de existência de um triângulo
        if (lado1 + lado2 > lado3) and \
           (lado1 + lado3 > lado2) and \
           (lado2 + lado3 > lado1):
            print("Os valores podem formar um triângulo válido.")
        else:
            print("Os valores NÃO podem formar um triângulo válido.")
    else:
        print("Para formar um triângulo, os valores dos lados devem ser positivos.")

except ValueError:
    print("Entrada inválida. Por favor, digite números válidos para os lados do triângulo.")
```

```python
#Crie um programa que leia a energia de um personagem (de 0 a 100) e classifique:
	#0: Morto
	#	1 a 30: Crítico
	# 31 a 70: Normal
	# 71 a 100: Saudável

energia_str = input("Por favor, digite a energia do personagem (de 0 a 100): ")

try:
    energia = int(energia_str)

    if energia == 0:
        classificacao = "Morto"
    elif 1 <= energia <= 30:
        classificacao = "Crítico"
    elif 31 <= energia <= 70:
        classificacao = "Normal"
    elif 71 <= energia <= 100:
        classificacao = "Saudável"
    else:
        classificacao = "Valor inválido. A energia deve estar entre 0 e 100."

    print(f"O personagem com {energia} de energia está classificado como: {classificacao}")

except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro para a energia.")
```

```python
#Crie um programa que leia a média final de um aluno e classifique:
	#	Média maior ou igual a 7: Aprovado
	#	Média entre 5 e 6.9: Recuperação

media_str = input("Por favor, digite a média final do aluno: ")

try:
    media = float(media_str)

    if media >= 7:
        classificacao = "Aprovado"
    elif 5 <= media <= 6.9:
        classificacao = "Recuperação"
    else:
        classificacao = "Reprovado (ou média fora das classificações especificadas)"

    print(f"O aluno com média {media:.1f} está: {classificacao}")

except ValueError:
    print("Entrada inválida. Por favor, digite um número para a média.")
```

```python
#Crie um programa que leia o salário de um funcionário e sua pontuação de desempenho. O programa deve calcular um bônus de acordo com as regras:
	#	Pontuação menor que 5: sem bônus
	#	Pontuação de 5 a 7: bônus de 5% do salário
	#	Pontuação de 8 a 9: bônus de 10% do salário
	#	Pontuação igual a 10: bônus de 15% do salário

#Ao final, mostre o valor do bônus e o salário final com o bônus aplicado.Média menor que 5: Reprovado

salario_str = input("Por favor, digite o salário do funcionário: R$")
pontuacao_str = input("Por favor, digite a pontuação de desempenho do funcionário (0 a 10): ")

try:
    salario = float(salario_str)
    pontuacao = int(pontuacao_str)

    bonus_percentual = 0.0

    if pontuacao < 5:
        bonus_percentual = 0.0 # Sem bônus
    elif 5 <= pontuacao <= 7:
        bonus_percentual = 0.05 # 5% de bônus
    elif 8 <= pontuacao <= 9:
        bonus_percentual = 0.10 # 10% de bônus
    elif pontuacao == 10:
        bonus_percentual = 0.15 # 15% de bônus
    else:
        print("Pontuação inválida. A pontuação deve estar entre 0 e 10.")
        bonus_percentual = -1 # Sinaliza erro para não calcular bônus

    if bonus_percentual != -1:
        valor_bonus = salario * bonus_percentual
        salario_final = salario + valor_bonus

        print(f"\nSalário base: R${salario:.2f}")
        print(f"Pontuação de desempenho: {pontuacao}")
        print(f"Bônus aplicado: {bonus_percentual * 100:.0f}%")
        print(f"Valor do bônus: R${valor_bonus:.2f}")
        print(f"Salário final com bônus: R${salario_final:.2f}")

except ValueError:
    print("Entrada inválida. Por favor, digite números válidos para o salário e a pontuação.")
```
